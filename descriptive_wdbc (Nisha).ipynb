```python
# Data cleaning 
#Identify the rows that contain the duplicates

# calculate duplicates
duplicate = df.duplicated()


print(duplicate.any())

#Search for Inconsistent data entry for output variables 
diagnosis = df['Diagnosis'].unique()
print(diagnosis)


# drop unnecessary column
df1 = df.drop(columns=(['ID','Diagnosis']))
print(df1.shape)

# search outlier using IQR score 
q1 = df1.quantile(0.25)
q3 = df1.quantile(0.75)
IQR = q3 - q1
print(IQR)

ul = q3 + (1.5*IQR)
print(ul)
print(df1 > ul)
print((df1 > ul).sum())

ll = q1 - (1.5*IQR)
print(df1 < ll)
print((df1 < ll).sum())

# Search for missing values
missing_value = df.isnull().sum()
print(missing_value)
```
